```python
def detect_outliers(data, x):
    Q1 = data[x].describe()['25%']
    Q3 = data[x].describe()['75%']
    IQR = Q3-Q1
    return data[(df[x] < Q1-1.5*IQR) | (data[x] > Q3+1.5*IQR)]
```
